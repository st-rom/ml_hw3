```python
def thresholding_algo(y, lag, threshold, influence):
    signals = np.zeros(len(y))
    filteredY = np.array(y)
    avgFilter = [0]*len(y)
    stdFilter = [0]*len(y)
    avgFilter[lag - 1] = np.mean(y[0:lag])
    stdFilter[lag - 1] = np.std(y[0:lag])
#     print(stdFilter)
    for i in range(lag, len(y)):
        if abs(y[i] - avgFilter[i-1]) > threshold * stdFilter [i-1]:
            if y[i] > avgFilter[i-1]:
                signals[i] = 1
            else:
                signals[i] = -1

            filteredY[i] = influence * y[i] + (1 - influence) * filteredY[i-1]
        else:
            signals[i] = 0
            filteredY[i] = y[i]
        avgFilter[i] = np.mean(filteredY[(i-lag):i+1])
        stdFilter[i] = np.std(filteredY[(i-lag):i+1])
#     signals = np.where(signals==1, 0, signals)
    print(pd.DataFrame({'y': y, 'fY': filteredY, 'avg': avgFilter, 'std': stdFilter, 'signals': signals})[4:80])
    return dict(signals = np.asarray(signals),
                avgFilter = np.asarray(avgFilter),
                stdFilter = np.asarray(stdFilter),
               filteredY = filteredY)
data1 = pd.read_csv("class_1/dir_1/raw_data/rd_2.csv", header=None)
print(pd.read_csv("class_1/dir_1/frames/f_2.csv", header=None))
y = [(data1[0][i] + data1[1][i]) / 2 for i in range(len(data1))]

lag = 9
threshold = 5
influence = 0.05

result = thresholding_algo(y, lag=lag, threshold=threshold, influence=influence)
pylab.subplot(211)
pylab.plot(np.arange(len(y)), y, color="cyan", lw=2)
pylab.plot(np.arange(len(y)),
           result["filteredY"], color="red", lw=2)
pylab.plot(np.arange(len(y)),
           result["avgFilter"], color="g", lw=2)
pylab.subplot(212)
pylab.step(np.arange(1, len(y)+1), result["signals"], color="red", lw=2)
len(count1(result["signals"], -1))
```

```python
def count1(res, ind):
    res = list(res)
    ranges = []
    i = 0
    while ind in res[i:]:
        i = res.index(ind, i)
        if 0 in res[i:]:
            j = res.index(0, i)
        else:
            j = len(res) - 1
        ranges.append((i, j))
        i = j
    return ranges
```
